# Reparto
Para obtener los artistas más populares del momento he realizado webscrapping en la lista de los 50 artistas más populares del momento en la web de IMDb. He recogido nombre, sexo, papel más actual y breve descripción de la persona.

El objetivo es crear un dataframe con los artistas del momento


```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.imdb.com/search/name/?match_all=true&ref_=rlm"
response = requests.get(url)
html= response.content
soup=bs(html, "lxml")
```

```python
#Webscrapping de los artistas
reparto=soup.find_all("h3", class_="lister-item-header")
```

```python
#Webscrapping del sexo del artista y su papel 
sexo=soup.find_all("p",class_="text-muted text-small")
```

```python
#Webscrapping de la descripción
description=soup.find_all("p")
description
```

```python
descript_pop = description[1].get_text()
#en los índices pares se encuentra el sexo/papel, en los índices impares la descripción
```

```python
new_lst = []
def odd_indices(lst): #función para sacar en una nueva lista las descripciones de cada artista

  for index in range(1, len(lst), 2):
    new_lst.append(lst[index])
  return new_lst

odd_indices(description)
new_lst
```

```python
descript_pop = new_lst[0].get_text()
des = (' '.join(descript_pop.split()))
des_title = des[len(str(0)):0]
print(descript_pop)
```

```python
imdb = []

for index in range(0,len(reparto)):
        reparto_pop = reparto[index].get_text()
        nombre = (' '.join(reparto_pop.split()).replace('.', ''))
        nombre_title = nombre[len(str(index))+1:]
    
        reparto_pop = reparto[index]

        descript_pop = new_lst[index].get_text()
        des = (' '.join(descript_pop.split()))
        
        sexo_pop = sexo[index].get_text()
        sex = (' '.join(sexo_pop.split()))
        
        data = {"Nombre": nombre_title,
            "Sexo/Papel": sex,
            "Descripción": des}
        imdb.append(data)
```

```python
df = pd.DataFrame(imdb)
df
```
